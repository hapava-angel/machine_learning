```python
import nltk
nltk.download('punkt')
import nltk.tokenize
import timeit
import spacy
```

```python
data = open('Les_Miserables-Victor_Hugo.txt', 'r', encoding='utf-8')
book = data.read()
book
```

Лексическое разнообразие текста:

```python
tokens = nltk.tokenize.RegexpTokenizer('\w+')
a = tokens.tokenize(book)

tokens_ = [x for x in [t.translate(a).lower() for t in a] if len(x) > 0]

text = nltk.Text(tokens_)
lexical_diversity = (len(set(text)) / len(text)) * 100
print(lexical_diversity)
```

Работа с параметрами RegexpTokenizer(3 блока).
Особенности(плюсы): Разделяет текст на токены с использованием регулярных выражений. Гибкий и позволяет определить собственные правила для токенизации. Много параметров можно фильтроваь (они прописаны чере # в блоках ниже). Скорость выполнения зависит от задачи, но токенизация совершается довольно быстро.
Минусы: Требует хорошо продуманных регулярных выражений для оптимальной работы.

```python
start_time = timeit.default_timer()

pattern = '[A-Z]\w+'
regexp = nltk.tokenize.RegexpTokenizer(pattern)
print(regexp.tokenize(book))

print(timeit.default_timer() - start_time)

#Можно задать параметры, какие именно слова выводить в токенах, например, здесь все слова с заглавной буквы, или все знаки препиания определенного типа (следующий блок), или вообще все символы
```

```python
start_time = timeit.default_timer()

pattern = '[A-Z]\w+'
regexp = nltk.tokenize.RegexpTokenizer(':')
print(len(regexp.tokenize(book)), regexp.tokenize(book))

print(timeit.default_timer() - start_time)

#также кол-вом точек можно обозначить кол-во символов, вписанных в один токене
```

```python
start_time = timeit.default_timer()

regexp = nltk.tokenize.RegexpTokenizer('\s+', gaps=True)  #логический параметр, определяющий, будет ли токенизатор искать сами токены (False) или разделители между ними (True). При gaps = True все пустые символы отброшены (автоматически использовано discard_empty)
print( regexp.tokenize(book))

print(timeit.default_timer() - start_time)
```

```python
start_time = timeit.default_timer()

whitespace = nltk.tokenize.WhitespaceTokenizer()
print(whitespace.tokenize(book))

print(timeit.default_timer() - start_time)
```

WhitespaceTokenizer разделяет текст на токены по пробелам, учитывая пунктуацию. Прост в использовании, обрабатывает текст с учетом пробелов и знаков препинания. #делит по словам, без пробелов, знаки препинания лепяться к словам
Минусы: Не учитывает специфические случаи, такие как аббревиатуры или смешанные слова.

```python
start_time = timeit.default_timer()

blank = nltk.BlanklineTokenizer()
print(blank.tokenize(book))

print(timeit.default_timer() - start_time)
```

BlanklineTokenizer разделяет текст по пустым строкам. Pазбивает текст, рассматривая в качестве разделителя любую последовательность пустых строк (содержащих только символы пробела или табуляции). Для этого используется регулярное выражение ‘\s*\n\s*\n\s*’. Просто использовать.
Минусы: Не подходит для текстов без пустых строк.

```python
start_time = timeit.default_timer()

word = nltk.WordPunctTokenizer()
print(word.tokenize(book))

print(timeit.default_timer() - start_time)
```

WordPunctTokenizer разделяет текст на токены, учитывая пунктуацию. Обрабатывает текст с учетом знаков препинания, подходит для большинства языков.
Минусы: Может создавать лишние токены для сложных случаев, таких как сокращения.

```python
start_time = timeit.default_timer()

sexpr = nltk.SExprTokenizer()
print(len(sexpr.tokenize(book)))
print(sexpr.tokenize(book))

print(timeit.default_timer() - start_time)
```

SExprTokenizer: Разделяет текст на токены, используя синтаксические выражения. Полезен для обработки структурированных данных. Довльно простой.
Минусы: Ограничен

```python
start_time = timeit.default_timer()

space = nltk.SpaceTokenizer()
print(space.tokenize(book))

print(timeit.default_timer() - start_time)

#Разбивает строку на токены, используя в качестве разделителя пробел. Результат работы этого метода полностью совпадает с результатом метода s.split(‘ ‘).
```

SpaceTokenizer pазделяет текст на токены по пробелам. Прост в использовании, быстрый.
Минусы: Не учитывает пунктуацию и спецсимволы.

```python
start_time = timeit.default_timer()

tab = nltk.TabTokenizer()
print(tab.tokenize(book))

print(timeit.default_timer() - start_time)

#Разбивает строку на токены, используя в качестве разделителя символ табуляции, что совпадает с применением метода s.split(‘\t’).
```

TabTokenizer разделяет текст на токены по символу табуляции что совпадает с применением метода s.split(‘\t’). Прост в использовании, хорош для обработки данных в формате таблиц.
Минусы: Ограничен в использовании для других типов текста.

```python
start_time = timeit.default_timer()

line = nltk.LineTokenizer()
print(line.tokenize(book))

print(timeit.default_timer() - start_time)
```

LineTokenizer pазделяет текст на токены по строкам, используя в качестве разделителя символ новой строки (при необходимости может отбрасывать пустые строки). Его применение похоже на применение метода s.split(‘\n’). Удобен для обработки структурированных текстов.
Минусы: Не учитывает предложения внутри строк.

```python
start_time = timeit.default_timer()

legality = nltk.LegalitySyllableTokenizer(book)
print(legality.tokenize(book))

print(timeit.default_timer() - start_time)
```

LegalitySyllableTokenizer разделяет текст на слоги с учетом языковых правил (можно использовать большую выборку языков), что начала и окончания слогов (не включая гласный звук) допустимы только в том случае, если они встречаются в качестве начал или окончаний слов в языке.Делит на слоги лучше, чем SyllableTokenizer.
Минусы: Может быть сложно обрабатывать разные языки и диалекты. Не удаляет символы \t, \n и тп

```python
start_time = timeit.default_timer()

syllabel = nltk.SyllableTokenizer()
print(syllabel.tokenize(book))

print(timeit.default_timer() - start_time)

#Это еще один токенизатор для разбиения слов на слоги. Его работа основана на принципе последовательности звучания (Sonority Sequencing Principle, SSP). 
```

Минусы: Не всегда подходит для всех языков, может быть не точным: объеденил в одном токене последний слог предыдущего слова, пробел и первый слог следующего - 'r o'

MWETokenizer (Multi-Word Expression Tokenizer): Идентифицирует и объединяет множественные слова в один токен. Работает только с другим токенайзером, например, на базе WordPunctTokenizer, можно выделить устойчивые выражения/именованные сущности, например выводить вместе имя и фамилию героя, как я это сделла, задва конкретный параметр; иначе просто делит посимвольно. 
Судя по уменьшившимуся кол-ву токенов можно заметить, сколько раз 'Jean_Valjean' является связанной конструкцией. Полезен для обработки фраз и выражений.
МИнусы: Требует заранее определенного списка множественных слов.

```python
start_time = timeit.default_timer()

tok = nltk.ToktokTokenizer()
print(tok.tokenize(book))

print(timeit.default_timer() - start_time)
```

ToktokTokenizer разделяет текст на токены, учитывая пунктуацию. Точен и быстр в обработке текста с учетом знаков препинания. Разделяет соседние знаки препинания в разные токены: ',', "'"; Но знаки, заканчивающие предложения прилепляет к предыдущему слову. В итоговой строке удаляются начальные и конечные пробелы, а сама строка преобразуется в кодировку Unicode: '&#9;'
Минусы: Может не всегда правильно обрабатывать текст с нетипичными символами.

TreebankWordTokenizer использует стандартные правила токенизации Treebank. Широко используется в обработке текста для многих языков, учитывает пунктуацию и сложные случаи. Разбивает на токены: слова и знаки препинания, сложные лова оставлет в одном токене: 'broad-shouldered'; Если в тексте есть прямая речь, кавычка и первое слово ПР идут в один токен: "'Then"; разделяет сокращения, типичные для английского языка: 'you', "'re"; если между словом и знаком препиания есть пробле, то он разделяет их  в разные токины, если нет - то в один.
Минусы: Может некорректно обрабатывать некоторые специфические случаи.
