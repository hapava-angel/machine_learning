```python
data = open('Les_Miserables-Victor_Hugo.txt', 'r', encoding='utf-8')
book = data.read()
book
```

```python
tokens = nltk.tokenize.RegexpTokenizer('\w+')
a = tokens.tokenize(book)

tokens_ = [x for x in [t.translate(a).lower() for t in a] if len(x) > 0]

text = nltk.Text(tokens_)
lexical_diversity = (len(set(text)) / len(text)) * 100
print(lexical_diversity)
```
